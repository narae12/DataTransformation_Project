#데이터 합치기
cars.csv와 brand.csv의 데이터 합치기 [힌트] cars.csv의 title 변수로부터 브랜드 데이터를 추출할 수 있습니다.

```python
#car_df의 타이틀을 소문자로 저장 후, 띄어쓰기를 기준으로 앞 부분만 저장하기

car_df['title'] = car_df['title'].str.lower()
car_df['title'] = car_df['title'].str.split().str[0]
car_df.head()
```

```python
#title 기준으로 합치기
car_df = pd.merge(left = car_df, right=brand_df, on='title', how='left')
car_df.head()
```

---
#카테고리형 변수 숫자로 바꾸기
카테고리형 변수를 숫자 형태로 변환하기
[힌트] Engine, Emission Class 변수에 주목하세요!

```python
car_df.info()
```

```python
#Engine 의 L를 없앤 후 숫자로 변환하기
car_df['Engine'] = car_df['Engine'].str.replace('L', '').astype('float')
```

```python
#Emission Class 의 Euro를 없앤 후 숫자로 변환하기
car_df['Emission Class'] = car_df['Emission Class'].str.replace('Euro', '').astype('float')
```

```python
car_df.head()
```

---
#결측치 처리하기
[힌트] Service history 변수의 결측치는 'Unknown'으로 채우고, 결측치 변수가 일정 개수 이상 포함된 데이터는 제거합니다.

```python
#결측치 확인하기
car_df.isnull().sum()
```

```python
#Service history 변수의 결측치는 'Unknown'으로 채우기
car_df['Service history'] = car_df['Service history'].fillna('Unknown')
```

```python
car_df.isna().mean()
```

```python
#Previous Owners 는 결측치가 많으니까 결측치데이터는 삭제하기
car_df = car_df.dropna(subset = ['Previous Owners'])
```

```python
#엔진은 평균값으로 대체하기
car_df['Engine'] = car_df['Engine'].fillna(car_df['Engine'].mean())
```

```python
car_df.isna().mean()
```

```python
#그냥 결측치 있는거 다 지워버리기
car_df = car_df.dropna()
```

```python
car_df.isna().sum()
```

```python
car_df.count()
#데이터가 많이 줄었지만 어쩔 수 없음....
```
